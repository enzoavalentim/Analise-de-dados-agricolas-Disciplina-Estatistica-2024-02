## **2. Product Name (Nome do Produto)**
- **Tipo de variável:** Categórica (Texto)
- **Análise Descritiva:**
  - **Moda:** Identificar o produto mais vendido.
  - **Distribuição por produto:** Verificar produtos com vendas significativas e identificar os menos vendidos.

```python
ContagemProdutos = dados['product_name'].value_counts()

ProdutoMaisVendido = ContagemProdutos.idxmax()
ProdutoMenosVendido = ContagemProdutos.idxmin()

QuantidadeProdutoMaisVendido = ContagemProdutos.max()
QuantidadeProdutoMenosVendido = ContagemProdutos.min()

print(f"O produto mais vendido é: {ProdutoMaisVendido} e aparece {QuantidadeProdutoMaisVendido} vezes.")
print(f"O produto menos vendido é: {ProdutoMenosVendido} e aparece {QuantidadeProdutoMenosVendido} vezes.")
```

O produto mais vendido é o Cheese e aparece 464 vezes e o menos vendido foi o Carrots que aparece apenas 215 vezes. 

## **3. Category (Categoria)**
- **Tipo de variável:** Categórica (Texto)
- **Análise Descritiva:**
  - **Contagem por categoria:** Número de produtos em cada categoria.
  - **Moda:** Categorias com maior demanda (baseado em vendas).
  - **Visualização:** Gráfico de barras ou pizza para analisar a distribuição entre categorias.

```python
import matplotlib.pyplot as plt

NumeroDeProdutosPorCategoria = dados.groupby('category')['product_name'].nunique()

print("Numero de produtos por categoria\n")
print(NumeroDeProdutosPorCategoria)

plt.figure(figsize=(10, 6))  
NumeroDeProdutosPorCategoria.plot(kind='bar', color='skyblue')

plt.title('Número de Produtos por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Número de Produtos Distintos')
plt.xticks(rotation=45, ha='right')  

plt.tight_layout()  
plt.show()


CategoriaComMaiorDemanda = dados['category'].mode()
print(f"\nCategoria com maior demanda de vendas: {CategoriaComMaiorDemanda}")
```

Temos 5 categorias de produtos onde as categorias Fruits e Vegetables estão empatadas como as categorias com maior variedade de produtos. 
A categoria com maior demanda de vendas e Fruits

## **4. Price per kg (Preço por Quilograma)**
- **Tipo de variável:** Numérica (Contínua)
- **Análise Descritiva:**
  - **Média e Mediana:** Determinar o preço médio e verificar simetria na distribuição.
  - **Mínimo e Máximo:** Identificar preços extremos (outliers).

```python
import matplotlib.pyplot as plt
import seaborn as sns

MediaPrecos = dados['price_per_kg'].mean()
MedianaPreco = dados['price_per_kg'].median()

print(f"Média: {MediaPrecos:.2f}")
print(f"Mediana: {MedianaPreco:.2f}")

plt.figure(figsize=(10, 6))
sns.histplot(dados['price_per_kg'], kde=True, bins=20, color='skyblue', label='Distribuição')

plt.axvline(MediaPrecos, color='red', linestyle='--', label='Média')
plt.axvline(MedianaPreco, color='green', linestyle='-', label='Mediana')

plt.title('Distribuição de Price per kg')
plt.xlabel('Price per kg')
plt.ylabel('Frequência')
plt.legend()

plt.tight_layout()
plt.show()
```

```python

Q1 = dados['price_per_kg'].quantile(0.25)  # Primeiro quartil
Q3 = dados['price_per_kg'].quantile(0.75)  # Terceiro quartil
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

plt.figure(figsize=(8, 6))
sns.boxplot(data=dados, y='price_per_kg', color='skyblue', width=0.4)

# Adicionando limites no gráfico
plt.axhline(limite_inferior, color='orange', linestyle='--', label='Limite Inferior')
plt.axhline(limite_superior, color='green', linestyle='--', label='Limite Superior')

# Adicionando títulos e rótulos
plt.title('Boxplot de Price per kg com Limites', fontsize=14)
plt.ylabel('Price per kg', fontsize=12)
plt.legend()

plt.tight_layout()
plt.show()

print(f"Limite Superior {limite_superior}")
```

Os valores que estão acima do limite superior (11.795) são considerados outliers

```python
#análise descritiva básica

descritivas = dados.describe().round(decimals=2)

descritivas = descritivas.rename(index={
    'count': 'Contagem',
    'mean': 'Média',
    'std': 'Desvio Padrão',
    'min': 'Mínimo',
    '25%': '1º Quartil',
    '50%': 'Mediana',
    '75%': '3º Quartil',
    'max': 'Máximo'
})

tabela_descritiva = descritivas.to_string()
print(tabela_descritiva)
```

```python
# Calcular a matriz de correlação

dados_numericos = dados.select_dtypes(include=['number'])

tabela_descritiva_com_correlacao = dados_numericos.corr()


pd.set_option('display.max_rows', None)

print("Tabela de Correlação:")
display(tabela_descritiva_com_correlacao)
```

```python
# Identificar as correlações mais fortes

abs_matriz_correlacao = tabela_descritiva_com_correlacao.abs()

melhores_correlacoes = abs_matriz_correlacao.unstack().sort_values(ascending=False)
melhores_correlacoes = melhores_correlacoes[melhores_correlacoes != 1]

print("Top 10 Melhores Correlações:")
print(melhores_correlacoes.head(10))
```
